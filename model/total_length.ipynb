```python
# import module
import cv2, os, glob, random
import datetime
import json
import math
```

```python
video_file_path = []
file_name_with_len = []

for file in os.listdir(os.getcwd()):
    if file.endswith('.ipynb'):
        continue
    video_file_path.append(os.getcwd() + '\\' + file)

for i in range(0, len(video_file_path)):
    file_total_num = 0
    for j in os.listdir(video_file_path[i]):
        file_total_num = len([entry for entry in os.listdir(video_file_path[i]) if os.path.isfile(os.path.join(video_file_path[i], entry))])
        class_label = video_file_path[i].split('\\')[-1]
        data = cv2.VideoCapture(video_file_path[i] + '\\' + j)
        # count the number of frames
        frames = data.get(cv2.CAP_PROP_FRAME_COUNT)
        fps = data.get(cv2.CAP_PROP_FPS)

        # calculate duration of the video
        seconds = round(frames / fps)
        video_time = datetime.timedelta(seconds=seconds)
        print(f"duration in seconds: {seconds}")
        print(f"video time: {video_time}")
        print('class_label', class_label)

        file_name_with_len.append({
            'vid_folder': class_label,
            'vid_num': file_total_num,
            'vid_name': j,
            'vid_len': video_time,
            'vid_second': seconds
        })

print(file_name_with_len)


# Serializing json
json_object = json.dumps(file_name_with_len, indent=4, default=str, ensure_ascii=False)
 
# Writing to sample.json
with open("data.json", "w", encoding="utf8") as outfile:
    outfile.write(json_object)
        
print('last', json_object)
```
